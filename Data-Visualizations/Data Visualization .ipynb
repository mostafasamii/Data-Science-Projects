# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### (1) Simple line chart

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gpd = [300.2, 543.3, 1075.9, 2826.5, 5979.6, 10289.7, 14958.3]
```

```python
#create a line char, years on x-axis and gpd on y-axis
plt.plot(years, gpd, color='green', marker='o', linestyle='solid')
# adding title
plt.title('Nominal GPD')
#add a label to the y-axis
plt.ylabel('Billions of $')

plt.show()
```

### (2) Bar chart

Simple Bar chart plot

```python
movies = ['Annie Halle', 'Ben-Hur', 'Casablance', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
```

```python
#plot bars with left x-coordinates [0,1,2,3,4] with the height [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("My favorite movies")
plt.ylabel("# of Academy Awards")

#label x-axis with movies name at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

* ------------------------------------------------------

Also Bar chart can be a good choice for plotting histograms of bucketed numeric values, in order to visualy explore how the values are distributed

```python
from collections import Counter
grades = [83, 95, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
#Bucket grades by decile, but put 100 in with the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()],    # shift bars right by 5
       histogram.values(),                  # give each bar its correct height
       10,                                  # give each bar a width of 10
       edgecolor=(0,0,0))                   # black edges for bars

plt.axis([-5, 105, 0, 5])                   # x-axis from -5 to 105 and y-axis from 0 to 5
plt.xticks([10*i for i in range(11)])       # x-axis labels at 0,10, .............100
plt.xlabel("decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam 1 grades")
plt.show()
```

* ----------------------------------------------------------------------------------------

Bar chart with misleading y-axis

```python
mentions = [500, 502]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# if times I head someone say 'Data Science'")

# if you don't do the following matplotlib will label the x-axis 0,1
# and then add a +2.016e3 off in the corner (bad matplotlib)
plt.ticklabel_format(useOffset=False)

#misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Loot at the 'Huge' Increase!")
plt.show()
```

The same chart with a nonmisleading y-axis

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# if times I head someone say 'Data Science'")

# if you don't do the following matplotlib will label the x-axis 0,1
# and then add a +2.016e3 off in the corner (bad matplotlib)
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("not so huge anymore!")
plt.show()
```
